```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer("nlpai-lab/KURE-v1")
vec_dim = len(model.encode("dummy_text"))
print(f"모델 차원: {vec_dim}")
```

```python
query1 = "피로·눈 건조 및 소화 불균형·체지방 관리를 위한 1일1포 무향 영양제, 저분자히알루론산·비오틴·아연 함유, 합성향료·타르색소·알레르기유발성분 프리"
query2 = "저분자 히알루론산 120mg을 함유한 1일 1포 드링크 타입 이너뷰티로 피부 속 수분 보유력을 높여 보습과 탄력 개선을 돕습니다."
query3 = "가볍게 발리는 오일 텍스처와 스포이드 패키지로 얼굴과 바디에 소량씩 레이어링하기 좋은 제형입니다." 
query4 = "가벼운 오일 형태, 올리브 오일이 포함된"
```

```python
query_vectors = [model.encode(q).tolist() for q in [query1, query2, query3, query4]]
vector_array = [np.array(v).reshape(1, -1) for v in query_vectors]
```

```python
# 2. 유사도 계산
similarity = cosine_similarity(vector_array[0], vector_array[1])

print(f"유사도: {similarity[0][0]}")
```

```python
similarity = cosine_similarity(vector_array[2], vector_array[3])

print(f"유사도: {similarity[0][0]}")
```

```python
similarity = cosine_similarity(vector_array[1], vector_array[2])

print(f"유사도: {similarity[0][0]}")
```
